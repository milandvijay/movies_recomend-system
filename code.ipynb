## convertion function

```python
import ast
def convert (obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.keywords[1]
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.cast[0]
```

```python
# we only nee first three name only 
def convert3 (obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies
```

```python
movies.crew[1]
```

# fetching the director 

```python

def fetch_director (obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.overview[1]
```

# extracting the Overview column 

```python
movies['overview']=movies['overview'].apply(lambda x:x.split() )
```

```python
movies
```

# removing the space 

```python
movies['genres']=movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['cast']+movies['overview']+movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']= new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

## **Applying Algorithem sklearn**

#### **steming --> ['loving','lover','love'] ---> ['love','love','love']**

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors =cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
len(cv.get_feature_names_out())
```

## cosine algoirithem to find the distance between the Vectors 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])
```

```python
def recommend (movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distance = similarity[movie_index]
    movies_list =sorted(list(enumerate(distance)),reverse=True,key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        # print(i[0])
```

```python
recommend('Avatar')
```

```python
recommend('Batman')
```
